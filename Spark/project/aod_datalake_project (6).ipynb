```pyspark
from pyspark.sql import SparkSession
from pyspark.sql.functions import col , column
import os
import time
import datetime
from pyspark.sql import functions as F
from pyspark.sql.types import DateType
from pyspark.sql.functions import udf
```

```pyspark
spark = SparkSession.builder\
                     .config("spark.jars.packages","org.apache.hadoop:hadoop-aws:2.7.0")\
                     .getOrCreate()
```

```pyspark
#works with col
songPath = "s3://udacity-dend/song_data/A/B/C/TRABCEI128F424C983.json"
#songPath = "s3://udacity-dend/song_data/*/*/*/*.json"
song_df = spark.read.json(songPath) \
            .withColumn("artist_id", col("artist_id").cast("int")) \
            .withColumn("year", col("year").cast("int")) \
            .withColumn("artist_latitude", col("artist_latitude").cast("float")) \
            .withColumn("artist_longitude", col("artist_latitude").cast("float"))
            
```

```pyspark
song_df.printSchema()
```

## Log Table

```pyspark
 

logPath = "s3://udacity-dend/log_data/2018/11/2018-11-12-events.json"
#logPath = "s3://udacity-dend/log_data/*/*/*.json"
log_df = spark.read.json(logPath) \
            .withColumn("ts_str", col("ts").cast("string"))
       
    
```

```pyspark

time_df = log_df \
        .withColumn('dateColumn', F.to_timestamp((log_df['ts_str']/1000).cast('timestamp'), "yyyy-MM-dd hh:mm:ss"))
```

```pyspark
time_df.printSchema()
```

```pyspark
time_df.show(1)
```

### UDF

```pyspark
# UDF

from pyspark.sql.types import *
from pyspark.sql import SQLContext
from pyspark.sql.functions import udf
from pyspark.sql.types import DateType

get_timestamp = udf(lambda x: datetime.datetime.fromtimestamp(x/1000).strftime('%Y-%m-%d %H:%M:%S'))

log2_df = log_df.withColumn("date_time", get_timestamp(log_df.ts))
```

```pyspark
log2_df.printSchema()
```

```pyspark
log2_df.show(1)
```

```pyspark
log3_df = log_df.withColumn('dateColumn', F.to_timestamp((log_df['ts']/1000).cast('timestamp'), "yyyy-MM-dd hh:mm:ss"))
```

```pyspark
log3_df.printSchema()
```

```pyspark
log3_df.show(1)
```

## songplays

```pyspark
df_songplays = song_df.join(log_df, (log_df.artist == song_df.artist_name) & (log_df.song == song_df.title) )  \
            .select(log_df["ts"], log_df["userId"], log_df["level"], song_df["song_id"], song_df["artist_id"], log_df["sessionid"] \
                , log_df["location"] , log_df["userAgent"] )
```
